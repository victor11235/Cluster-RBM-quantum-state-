```python
import netket as nk
import json
from qutip import *
import numpy as np
import time
import multiprocessing as mp
from collections import OrderedDict
from pickle import dump
import os
import matplotlib.pyplot as plt
import scipy
from matplotlib import gridspec
from functools import reduce
plt.style.use('seaborn')
from scipy.stats import norm
```

```python
def ran():
    np.random.seed()
    a = np.random.random()
    return a

numRuns = 50000
hisIt = np.arange(numRuns)

# Cluster multiproccessing
ncpus = int(os.environ.get('SLURM_CPUS_PER_TASK',default=10))
pool = mp.Pool(processes=ncpus)
# Run Descent
results = []
for j in hisIt:
    result = pool.apply_async(ran, args=()) 
    results.append(result)
    
pool.close()
pool.join()
results_ran = [r.get() for r in results]
    

print(results_ran)
    
```

```python
bins = np.arange(0, 1.0, 0.01)
plt.hist(results_ran, bins=bins)
plt.xlabel('Random Number Value')
plt.legend()
plt.show()
```
